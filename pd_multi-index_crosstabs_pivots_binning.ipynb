Pandas Cross Tab – pd.crosstab()
For all intensive purposes, you can think of Pandas Crosstab as the same things as Pandas Pivot Table.

When to use pd.crosstab(): When you are starting with non-DataFrame based data. This could be a list of lists or dictionaries.

When to use pd.pivot_table(): When you are starting with a Dataframe.

```python
pd.crosstab(index=stocks.Symbol, columns=stocks.Date, values=stocks.Close, aggfunc=lambda x: x.min()**2, margins=True, margins_name='Totals', rownames=['Stocks'], colnames=['Dates'])
```

```python
pd.cut(stocks.Volume, 3, right=False, precision=0, labels=[22,33,44], include_lowest=True)
```

```python
pd.cut(stocks.Volume, 3, right=False, precision=0, labels=[22,33,44], include_lowest=True, retbins=True)
```

```python
pd.cut(stocks.Volume, 4, right=False, include_lowest=False, labels=['Q1', 'Q2', 'Q3', 'Q4'], precision=1)
```
