```python
df.select([pl.min('mpg').alias('minimum'), pl.max('mpg').alias('maximum')], )
```

```python
df.select(
    pl.when(pl.col('mpg')>35)\
    .then('Amazing')\
    .otherwise(pl.when(pl.col('mpg')>25).then('Good').otherwise('avg')).alias('bins'))
```

```python
df.with_columns([pl.when(pl.col('mpg')>35).then('Good').otherwise(pl.when(pl.col('mpg')>20).then('average').otherwise('ok')).alias('sss')])
```

```python
df = df.with_columns([pl.when(pl.col('mpg')>35).then('Good').otherwise(pl.when(pl.col('mpg')>20).then('average').otherwise('ok')).alias('sss')])
```
