```python
values = np.random.randn(1000, 4)
index = pd.date_range('1/1/2023', periods=1000)

df = pd.DataFrame(values, index=index, columns=['A', 'B', 'C', 'D'])
df = df.cumsum()

df.plot()
```

```python
df.plot(subplots=True, figsize=(6,6));
```

```python
df.plot(subplots=True, layout=(2,2), figsize=(6,6), sharex=True, sharey=True);
```

```python
pd.plotting.autocorrelation_plot(df)
```

```python
plt.style.use('ggplot')
ax = df.plot()
pd.plotting.table(ax=ax, data=df.describe(), loc='upper right')
```

```python
pd.plotting.bootstrap_plot(df.A, size=50, samples=200)
```

```python
pd.plotting.parallel_coordinates(df.iloc[0:10], class_column='A')
```

```python
df.A.plot.kde()
```

```python
pd.plotting.andrews_curves(df.iloc[:10], class_column='A')
```
